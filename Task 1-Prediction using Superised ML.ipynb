```python
#IMPORTING LIBRARIES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from scipy import stats
```

```python
#READING GIVEN DATA
df = pd.read_csv('http://bit.ly/w-data')
df
```

```python
#PLOTING SCATTER GRAPH OF GIVEN DATA
%matplotlib inline
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.scatter(df.Hours,df.Scores)
plt.show()
```

```python
#PLOTING THE REGRESSION LINE
Hours=list(df.Hours)
Scores=list(df.Scores)

slope, intercept, r, p, std_err = stats.linregress(Hours,Scores)

def myfunc(Hours):
  return (slope * Hours) + intercept

mymodel = list(map(myfunc, Hours))

plt.scatter(Hours,Scores)
plt.plot(Hours, mymodel)
plt.show()
```

```python
#CREATING REGRESSION MODEL
reg =linear_model.LinearRegression()
reg.fit(df[['Hours']],df.Scores)
```

```python
#CALCULATING SLOPE AND INTERCEPT OF LINE 
m=reg.coef_
print('Slope of the line is:' ,m)
c=reg.intercept_
print('The intercept of the line is:' ,c)
```

```python
#PREDICTED SCORE
p=myfunc(9.25)
print('The predicted scores:', p)
```
